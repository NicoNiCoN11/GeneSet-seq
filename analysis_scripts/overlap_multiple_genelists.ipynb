```python
import pandas as pd
import sys
sys.path.append('/home/jiguo/denovo_rpe1_scrnaseq/analysis_pipeline')
import functions as f
```

```python
list1 = f.load_gene_list('/home/jiguo/denovo_rpe1_scrnaseq/locations_of_genes/group1.txt', gene_index='Gene name')
list2 = f.load_gene_list('/home/jiguo/denovo_rpe1_scrnaseq/locations_of_genes/group2.txt', gene_index='Gene name')
```

```python
def find_common_genes(list1, list2):
    common_genes = set(list1).intersection(set(list2))
    return common_genes
common_genes = find_common_genes(list1, list2)
print(f"Number of common genes: {len(common_genes)}")
print("Common genes:")
for gene in common_genes:
    print(gene)
```

```python
def find_uncommon_genes(list1, list2):
    uncommon_genes = set(list1).difference(set(list2))
    return uncommon_genes
uncommon_genes = find_uncommon_genes(list1, list2)
print(f"Number of uncommon genes: {len(uncommon_genes)}")
print("Uncommon genes:")
for gene in uncommon_genes:
    print(gene)
```
